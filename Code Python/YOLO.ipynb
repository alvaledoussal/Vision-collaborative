```python
import cv2
import matplotlib.pyplot as plt

from utils import *
from darknet import Darknet
```

# Réseaux de neurones




```python

cfg_file = './cfg/yolov3.cfg'
weight_file = './weights/yolov3.weights'
namesfile = 'data/coco.names'

# Architecture
m = Darknet(cfg_file)

m.load_weights(weight_file)

class_names = load_class_names(namesfile)
```

```python
# Architecture du réseaux de neurones utilisés
m.print_network()
```


Il faut changer les dimensions des images à 416 x 416 x 3 pour qu'elles soient compatibles avec l'input size du réseau, pour cela on utilise la fonction`cv2.resize()`;

```python

plt.rcParams['figure.figsize'] = [24.0, 14.0]

img = cv2.imread('./images/wine.jpg')

original_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
resized_image = cv2.resize(original_image, (m.width, m.height))

# Display the images
plt.subplot(121)
plt.title('Image originale')
plt.imshow(original_image)
plt.subplot(122)
plt.title('image_bis')
plt.imshow(resized_image)
plt.show()
```

```python

iou_thresh = 0.4
nms_thresh = 0.6

# Detecter les objets dans les images 
boxes = detect_objects(m, resized_image, iou_thresh, nms_thresh)

# niveau de confiance
print_objects(boxes, class_names)

plot_boxes(original_image, boxes, class_names, plot_labels = True)
```
